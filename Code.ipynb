```python
bins=np.linspace(18,60,11)
labels=['18-22','23-26','27-30','31-34','35-38','39-42','43-46','47-50','51-55','56-59']
df['Age Group']=pd.cut(df['Age'],bins=bins,labels=labels)
df_age=df['Age Group'].value_counts().to_frame().reset_index()
df_age['Age Group']=pd.Categorical(df_age['Age Group'], categories=labels,ordered=True)
df_age=df_age.sort_values('Age Group')

plt.figure(figsize=(10,5))
sns.countplot(x=df['Age Group'], palette='viridis', hue=df['Age Group']);
plt.title('Age Distribution');
```

### Key Insights:

* The age group 23-26 has the smallest representation, with just over 400 individuals.
Overall, the age groups show relatively small variations in size.

```python
def float_to_Time(time_float):
    hours=int(time_float)
    mins=round((time_float-hours)*60)
    return f'{hours:02}:{mins:02}'

print(f'The earliest sleep start time in records: {float_to_Time(min(df['Sleep Start Time']))}')
print(f'The earliest sleep end time in records: {float_to_Time(min(df['Sleep End Time']))}')
print(f'The latest sleep start time in records: {float_to_Time(max(df['Sleep Start Time']))}')
print(f'The latest sleep end time in records: {float_to_Time(max(df['Sleep End Time']))}')
```

```python
fig,ax=plt.subplots(6,2,figsize=(35,30))
colors = ['#FF6347', '#4682B4', '#32CD32', '#FFD700', '#8A2BE2', '#FF1493', '#00CED1', '#FF4500', '#2E8B57', '#D2691E', '#20B2AA', '#F0E68C']
for i in range(4,15):
    bins=np.linspace(min(df[f'{df.columns[i]}']),max(df[f'{df.columns[i]}']),11)
    df[f'{df.columns[i]} Group'] = pd.cut(df[f'{df.columns[i]}'], bins=bins)
    color_index = (i - 4) % len(colors)
    new_df = df[f'{df.columns[i]} Group'].value_counts().to_frame().sort_values(f'{df.columns[i]} Group').reset_index()
    new_df[f'{df.columns[i]} Group']=pd.Categorical(new_df[f'{df.columns[i]} Group'], ordered=True)
    ax[(i-4)//2][(i-4)%2].plot(new_df[f'{df.columns[i]} Group'].astype(str),new_df['count'], color=colors[color_index])
    ax[(i-4)//2][(i-4)%2].set(title=f'{df.columns[i]} Distribution', ylabel='Count', xlabel=f'{df.columns[i]} Group')
    ax[(i-4)//2][(i-4)%2].grid(True)
    
plt.subplots_adjust(hspace=0.5)
fig.delaxes(ax[5][1])
    
```

#### 1. Sleep-Related Distributions

##### 1.1 Sleep Start Time Distribution
- The distribution of sleep start times shows significant variation, with no clear fixed pattern.
- Certain time ranges, particularly **9:00 - 10:00 PM** and **11:00 - 11:30 PM**, have higher counts.
- Frequent variations in sleep start time may impact sleep quality and circadian rhythm.

##### 1.2 Sleep End Time Distribution
- Wake-up times are more stable, with the highest frequency occurring around **5:00 - 6:00 AM**.
- After this peak, the count gradually declines, indicating that most people wake up around a fixed time.

##### 1.3 Total Sleep Hours Distribution
- The most common sleep duration falls between **6 - 8 hours**.
- A small portion of individuals sleep **less than 5 hours**, which could negatively impact health and productivity.

##### 1.4 Sleep Quality Distribution
- Sleep quality distribution remains relatively stable, with no significant variation among groups.
- Sleep quality scores range mostly between **5 - 8** on a 10-point scale.

#### 2. Lifestyle Factors

##### 2.1 Exercise (mins/day) Distribution
- Daily exercise time is widely distributed, with the highest concentration in the **20 - 40 minutes** range.
- Some individuals exercise for less than **10 minutes/day**, which may impact overall health and sleep quality.

##### 2.2 Caffeine Intake (mg) Distribution
- Caffeine consumption tends to decrease after exceeding **150 mg/day**.
- Some individuals consume more than **250 mg/day**, which could negatively affect sleep and stress levels.

##### 2.3 Screen Time Before Bed (mins) Distribution
- Screen time before bed varies significantly, with many individuals spending **60 - 120 minutes** on screens.
- Some individuals exceed **140 minutes**, which could negatively impact sleep quality.

#### 3. Work, Productivity, and Mental Health

##### 3.1 Work Hours (hrs/day) Distribution
- The average work duration falls between **6 - 9 hours/day**.
- Some individuals work **more than 10 hours/day**, which may correlate with higher stress levels and lower work-life balance.

##### 3.2 Productivity Score Distribution
- Productivity scores are relatively stable, with most values ranging from **5 - 8** on a 10-point scale.
- Some peaks suggest that either overwork or effective time management strategies play a role.

##### 3.3 Stress Level Distribution
- Stress levels tend to be higher in the **2 - 6** range on a 10-point scale.
- While some individuals have low stress levels, a significant portion experiences moderate to high stress.

##### 3.4 Mood Score Distribution
- Mood scores remain stable, with the majority falling between **6 - 8** on a 10-point scale.
- No major fluctuations indicate that mood remains relatively consistent across the dataset.

---

#### Conclusion
- Sleep duration and quality vary significantly, with some individuals sleeping less than 5 hours.
- High caffeine intake and excessive screen time before bed could negatively impact sleep.
- Stress levels and productivity scores show a stable pattern, though some individuals work excessively.
- Exercise habits are highly variable, with some individuals engaging in very little physical activity.

Overall, the data suggests that lifestyle factors, including sleep habits, screen time, caffeine intake, and work hours, may impact sleep quality, productivity, and mental health. Maintaining a consistent sleep schedule, limiting screen exposure before bed, reducing caffeine intake, and ensuring adequate rest and exercise can significantly improve overall well-being.


## Bivariate

### Analyze the relationship between sleep duration and productivity.

```python
sleep_duration_vs_productivity = df.groupby(['Total Sleep Hours Group', 'Productivity Score'], observed=True).size().reset_index(name='count')
sleep_duration_vs_productivity = pd.pivot_table(data=sleep_duration_vs_productivity, index='Productivity Score', columns='Total Sleep Hours Group', values='count',observed=False)
# sleep_duration_vs_productivity

productivity_vs_quality=df.groupby(['Sleep Quality','Productivity Score'],observed=True).size().reset_index(name='count')
productivity_vs_quality = pd.pivot_table(data=productivity_vs_quality,index='Productivity Score', columns='Sleep Quality', values='count',observed=True)
# productivity_vs_quality

```

### Examines how exercise, caffeine, and screen time affect sleep quality.

```python
fig,ax=plt.subplots(1,2,figsize=(10,4))
sns.heatmap(data=sleep_duration_vs_productivity,cmap='viridis',linewidths=1,ax=ax[0]);
sns.heatmap(data=productivity_vs_quality,cmap='coolwarm',linewidths=1,ax=ax[1]);
```

```python
fig,ax= plt.subplots(3,1,figsize=(10,20))

for i in range(8,11):
    new_df1=df.groupby(f'{df.columns[i]}')['Sleep Quality'].mean()
    new_df1=new_df1.to_frame().reset_index()
    sns.boxplot(data=new_df1, x=f'{df.columns[i]}', y='Sleep Quality',dodge=True,ax=ax[i-8])
    positions=np.linspace(0,len(new_df1),11)
    positions=list(positions)
    ax[i-8].set_xticks(positions)
    ax[i-8].set_xticklabels(positions);
    ax[i-8].grid(True)



```

### Identifies patterns in stress levels and mood based on sleep habits.


```python
fig, ax=plt.subplots(3,1,figsize=(10,15))
for i in range(8,11):
    new_df=df.groupby(f'{df.columns[i]} Group', observed=False)['Mood Score'].mean()
    new_df=new_df.to_frame().reset_index()
    new_df1=df.groupby(f'{df.columns[i]} Group', observed=False)['Stress Level'].mean()
    new_df1=new_df1.to_frame().reset_index()
    ax[i-8].plot(new_df[f'{df.columns[i]} Group'].astype(str), new_df['Mood Score'],label='Mood Score', marker='o');
    ax[i-8].plot(new_df1[f'{df.columns[i]} Group'].astype(str),new_df1['Stress Level'],label='Stress Level',marker='s');
    ax[i-8].grid(True)
    ax[i-8].legend()
    ax[i-8].set(title=f'Impact of {df.columns[i]} on Mood & Stress', xlabel=f'{df.columns[i]}',ylabel=f'Level of impact')
    a=df.groupby(f'{df.columns[i]}', observed=False)['Mood Score'].mean()
    positions=list(np.linspace(0,len(a),11))
    
    ax[i-8].set_xticks(list(np.linspace(0,9,11)))
    ax[i-8].set_xticklabels(positions);

plt.tight_layout()
```
